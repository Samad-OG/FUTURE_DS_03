```python
import pandas as pd
import plotly.graph_objects as go

# 1. Data Load (Excel file ke liye updated)
try:
    # Aapki file .xlsx hai isliye read_excel use karenge
    df = pd.read_excel('bank.xlsx')
    print("‚úÖ Excel Dataset Loaded Successfully!")

    # 2. Funnel Stages Calculation
    # Note: Dataset columns ke hisaab se calculation
    total = len(df)

    # Contacted leads (Unknown hata kar)
    contacted = len(df[df['contact'].str.lower() != 'unknown']) if 'contact' in df.columns else total * 0.8

    # Interested (Duration ke basis par)
    interested = len(df[df['duration'] > 200]) if 'duration' in df.columns else total * 0.4

    # Final Conversion (Deposit column check karein)
    target_col = 'deposit' if 'deposit' in df.columns else 'y'
    converted = len(df[df[target_col].str.lower() == 'yes'])

    stages = ["Total Leads", "Successfully Contacted", "High Interest", "Final Conversion"]
    values = [total, contacted, interested, converted]

    # 3. Professional Funnel Chart
    fig = go.Figure(go.Funnel(
        y = stages,
        x = values,
        textinfo = "value+percent initial",
        marker = {"color": ["#1abc9c", "#3498db", "#9b59b6", "#e74c3c"]},
        connector = {"line": {"color": "white", "width": 2}}
    ))

    fig.update_layout(
        title_text="üöÄ Marketing Funnel & Conversion Performance Analysis",
        template="plotly_dark"
    )
    fig.show()

    # Final Conversion Rate
    cr = (converted/total)*100
    print(f"\nüìä Overall Conversion Rate: {cr:.2f}%")

except Exception as e:
    print(f"‚ùå Abhi bhi error hai: {e}")
    print("üí° Tip: Left panel mein file par right-click karke 'Rename' karein aur check karein ki naam exactly 'bank.xlsx' hi hai.")
```
